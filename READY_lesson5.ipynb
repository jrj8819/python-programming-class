### 두가지 인터페이스

Matplotlib는 매트랩 스타일의 인터페이스와 객체지향 인터페이스 2가지를 지원한다.

#### 매트랩 인터페이스

```python

plt.figure() # 플롯 그림 생성

# 두개의 패널 중 첫 패널을 생성

plt.subplot(2, 1, 1) #(rows, columns, panel number)
plt.plot(x, np.sin(x))

```

```python

# 두번째 패널 생성
plt.subplot(2, 1, 2) #(rows, columns, panel number)
plt.plot(x, np.cos(x))

```

#### 객체지향 인터페이스

```python

# 전체 플롯그리드(fig), 두 객체의 배열(ax) 생성
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

## 간단한 라인 플롯

모든 Matplotlib 플롯은 그림(figure)와 축(axes)를 만드는 것으로 시작
- 그림(figure) : 축, 그래픽, 텍스트, 레이블을 표시하는 하나의 컨테이너(ex) 그림이 그려지는 종이)
- 축(axes) : 하나의 플롯 그래프( 영어에서 축을 의미하는 axis의 복수가 axes, 철자에 주의하자)
- axis : x축, y축을 의미

```python
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```

### 플롯 수정하기 : 선 색상과 스타일

플롯에서 선의 색상과 스타일을 수정할 때 plt.plot() 함수의 color 키워드를 사용

```python
plt.plot(x, np.sin(x-0), color='blue')
plt.plot(x, np.sin(x-1), color='g')
plt.plot(x, np.sin(x-2), color='0.75')
plt.plot(x, np.sin(x-3), color='#FFDD44')
plt.plot(x, np.sin(x-4), color=(0,1,0))
```

linestyle 키워드를 활용하면 선 스타일을 선택할 수 있다.

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

linestyle과 color코드를 합쳐 표현할 수도 있다.

```python
plt.plot(x, x+0, '-g')  # solid + green
plt.plot(x, x+1, '--c') # dashed + cyan
plt.plot(x, x+2, '-.k') # dashdot + black
plt.plot(x, x+3, ':r')  # dotted + red
```

단일문자로 표현되는 색상은 RGB(Red, Green, Blue), CMYK(Cyan(청록), Magenta(자홍), Yellow, BlacK)의 약어이다.

### 플롯 조정하기 : 축 경계

Matplotlib은 기본적으로 축 경계를 자동으로 설정한다. 축 경계를 조정하기 위해서는 plt.xlim(), plt.ylim() 매서드를 활용한다.

```python
plt.plot(x, np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

축을 반대로 표시해야 하면 인수의 순서를 바꾸자

```python
plt.plot(x, np.sin(x))
plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

plt.axis() 매서드는 [x축 최소값, x축 최대값, y축 최소값, y축 최대값]를 지정하는 리스트로 x, y축을 설정한다.

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5,1.5])
```

축에 대한 리스트 값 대신 plt.axis()의 매개변수로 
- 'tight'를 주면 플롯 주변의 경계를 밀착시킬 수 있고, 
- 'equal'을 지정하면 축의 단위를 똑같게 설정하여 보여준다.

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

## 오차 시각화하기

### 오차 막대

데이터의 결과의 시각화에서 오차를 효과적으로 보여주면 훨씬 더 완전한 정보를 전달 할 수 있다.
- 예시) 설문조사 오차, 천체물리 관측치

```python
x = np.linspace(0,10,50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='.k')
```

fmt은 선과 점의 모양을 제어하는 키워드로 라인과 산점도에서 활용했던 스타일과 색상의 약칭을 입력한다.

```python
plt.errorbar(x, y, xerr=dy, fmt='o', color='k', 
             ecolor='lightgray', elinewidth=3, capsize=5)
```

ecolor는 오류부분의 색상,
elinewidth는 오류 부분의 너비
capsize는 오류 경계부분의 사이즈

## 밀도 플롯과 등고선 플롯

3차원 데이터를 등고선이나 색으로 구분한 영역을 사용해 2차원으로 표시하는 것이 유용하다. matplotlib은 이를 위한 3가지 함수(plt.contour, plot.contourf)를 지원한다

### 3차원 함수 시각화 하기

다음 함수로 등고선 플롯을 그려보자

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10+y*x) * np.cos(x)
```

```python
x = np.linspace(0,4,3)
y = np.linspace(0,4,3)
X,Y = np.meshgrid(x,y) # 가로축 세로축을 이루는 점 좌표를 받아 사각형 영역을 이루는 조합 출력
```

```python
plt.scatter(X, Y, linewidths=10)
```

```python

x = np.linspace(0,5,50)
y = np.linspace(0,5,40)
X,Y = np.meshgrid(x,y) 

Z = f(X, Y)
```

plot.contour함수로 등고선 플롯을 만들어 보자

```python
plt.contour(X, Y, Z, colors='Black') # x, y는 플롯의 위치 z는 등고선의 높이를 의미
```

x,y,z, 다음에 간격수 20을 입력하여 데이터의 범위를 나누었다.
cmap 인수로 색상표를 지정해 선에 색을 입힐 수 있다.

```python
plt.contour(X,Y,Z,20,cmap='RdGy') 
```

색상표는 plt.cm 모듈에 정의 되어 있으므로 탭 완성 기능을 이용하여 색상을 쉽게 확인 가능하다.

plt.cm.Tab키

plt.contourf()함수는 선 사이의 공간안에 색을 체워 등고선 플롯을 그릴 수 있다.

```python
plt.contourf(X,Y,Z, 20, cmap='RdGy')
plt.colorbar()
```

plt.imshow()함수는 2차원 데이터 그리드를 이미지로 해석하여 부드러운 형태의 플롯을 그릴 수 있다.
다만
- plt.imshop()은 x, y 그리드 좌표를 받지 않으므로 플롯의 이미지 크기 extent를 직접 지정한다.
    - [x_min,x_max,y_min,y_max]

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
```

## 플롯 범례 변경하기

Matplotlib에서 범례의 모양과 위치를 맞춤 변경하는 것을 알아보자

```python
plt.style.use('classic')
x = np.linspace(0,10,1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sin')
ax.plot(x, np.cos(x), '--r', label='Cos')
ax.axis('equal')
leg = ax.legend()
```

legend()함수의 loc 키워드로 범례의 위치를 frameon 키워드로 테두리를 지정할 수 있다. 

```python

ax.legend(loc='lower left', frameon=False) # upper/lower left/right/center
fig
```

이외에 둥근 모서리박스 사용(fancybox), 투명도(framealpha), 음영(shadow), 박스내 여백(borderpad) 등을 변경할 수 있다.

```python
ax.legend(fancybox=True, framealpha=0.5, shadow=True, borderpad=1)
fig
```

## 서브 플롯

자료를 비교할 때 서로 다른 데이터 뷰를 나란히 비교하는 것이 유용하다. Matplotlib에서는 서브플롯(subplot)이라는 개념이 있다.


### 그리드 형태

plt.subplot(격자의 행, 격자의 열, 서브플롯의 인덱스)함수는 그리드(격자)안에 들어갈 여러 서브플롯을 생성한다.

```python
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5, 0.5, str((2,3,i)),fontsize=18, ha='center')
```

plt.subplots_adjust()함수의 hspace, wspace로 플롯 간 높이, 너비 사이의 간격을 조정할 수 있다

```python
plt.subplots_adjust(hspace=0.5, wspace=0.3)
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5, 0.5, str((2,3,i)),fontsize=18, ha='center')
```
