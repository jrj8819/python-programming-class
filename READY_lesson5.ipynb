### 두가지 인터페이스

Matplotlib는 매트랩 스타일의 인터페이스와 객체지향 인터페이스 2가지를 지원한다.

#### 매트랩 인터페이스

```python

plt.figure() # 플롯 그림 생성

# 두개의 패널 중 첫 패널을 생성

plt.subplot(2, 1, 1) #(rows, columns, panel number)
plt.plot(x, np.sin(x))

```

```python

# 두번째 패널 생성
plt.subplot(2, 1, 2) #(rows, columns, panel number)
plt.plot(x, np.cos(x))

```

#### 객체지향 인터페이스

```python

# 전체 플롯그리드(fig), 두 객체의 배열(ax) 생성
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

## 간단한 라인 플롯

모든 Matplotlib 플롯은 그림(figure)와 축(axes)를 만드는 것으로 시작
- 그림(figure) : 축, 그래픽, 텍스트, 레이블을 표시하는 하나의 컨테이너(ex) 그림이 그려지는 종이)
- 축(axes) : 하나의 플롯 그래프( 영어에서 축을 의미하는 axis의 복수가 axes, 철자에 주의하자)
- axis : x축, y축을 의미

```python
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
```

### 플롯 수정하기 : 선 색상과 스타일

플롯에서 선의 색상과 스타일을 수정할 때 plt.plot() 함수의 color 키워드를 사용

```python
plt.plot(x, np.sin(x-0), color='blue')
plt.plot(x, np.sin(x-1), color='g')
plt.plot(x, np.sin(x-2), color='0.75')
plt.plot(x, np.sin(x-3), color='#FFDD44')
plt.plot(x, np.sin(x-4), color=(0,1,0))
```

linestyle 키워드를 활용하면 선 스타일을 선택할 수 있다.

```python
plt.plot(x, x+0, linestyle='solid')
plt.plot(x, x+1, linestyle='dashed')
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

linestyle과 color코드를 합쳐 표현할 수도 있다.

```python
plt.plot(x, x+0, '-g')  # solid + green
plt.plot(x, x+1, '--c') # dashed + cyan
plt.plot(x, x+2, '-.k') # dashdot + black
plt.plot(x, x+3, ':r')  # dotted + red
```

단일문자로 표현되는 색상은 RGB(Red, Green, Blue), CMYK(Cyan(청록), Magenta(자홍), Yellow, BlacK)의 약어이다.

### 플롯 조정하기 : 축 경계

Matplotlib은 기본적으로 축 경계를 자동으로 설정한다. 축 경계를 조정하기 위해서는 plt.xlim(), plt.ylim() 매서드를 활용한다.

```python
plt.plot(x, np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

축을 반대로 표시해야 하면 인수의 순서를 바꾸자

```python
plt.plot(x, np.sin(x))
plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

plt.axis() 매서드는 [x축 최소값, x축 최대값, y축 최소값, y축 최대값]를 지정하는 리스트로 x, y축을 설정한다.

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5,1.5])
```

축에 대한 리스트 값 대신 plt.axis()의 매개변수로 
- 'tight'를 주면 플롯 주변의 경계를 밀착시킬 수 있고, 
- 'equal'을 지정하면 축의 단위를 똑같게 설정하여 보여준다.

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

### 플롯에 레이블 붙이기

플롯에 제목, 축 레이블(축이름), 간단한 범례를 추가할 수 있다. 

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve", fontsize=20, color='green', loc='center')
plt.xlabel("X")
plt.ylabel("sin(X)")
```

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve", fontsize=20, color='green', loc='center')
plt.xlabel("X", fontsize=15, color='blue')
plt.ylabel("sin(X)", fontsize=15, color='red')
```

범례를 만들려면 1) label 키워드를 사용해 각 라인의 레이블을 지정하고, 2) plt.legend() 메서드를 사용한다.

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), '-b', label='cos(x)')
plt.axis('equal')

plt.legend()
```

## 산점도(Scatter plot)

산점도는 점을 선으로 연결하는 대신 데이터를 개별적인 원 혹은 다른 기호로 표현한다

### plt.plot을 사용한 산점도

```python
x = np.linspace(0,10,30)
```

```python
y = np.sin(x)
```

```python
# https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.plot.html
```

```python
plt.plot(x,y,'^', color='black')
```

```python
rng = np.random.RandomState(0)
for marker in ['o','.',',','x','v','^','<','>','s','d']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker=%s" % marker)
    plt.legend(numpoints=1) # numpoints 키워드는 범례의 마커 표시 갯 수
    plt.xlim(0, 1.8)
```

```python
plt.plot(x, y, '-.^y') # 선 마커기호 노랑
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x, y, '-p', color='gray',
        markersize = 15, 
        linewidth = 4,
        markerfacecolor = 'white',
        markeredgecolor = 'gray',
        markeredgewidth = 2)
plt.ylim(-1.2, 1.2)
```

### plt.scatter를 활용한 산점도

plt.plot 함수와 비슷하게 사용할 수 있는 plt.scatter 함수가 있다.

```python
plt.scatter(x,y,marker='o')
```

plt.scatter는 plt.plot와 달리 각 점의 속성(크기, 표면 색상, 테두리, 색상 등)을 개별적으로 제어하거나 데이터에 매핑할 수 있다.

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100) # 평균이 0이고 표준편차가 1인 정규분포 값 100개
colors = rng.rand(100) # 0 ~ 1 사이 값 100개
sizes = 1000 * rng.rand(100)

plt.scatter(x,y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```

```python
import pandas as pd
iris = pd.read_csv('iris.csv')
```

```python
iris.head()
```

```python
def setColor(x):
    if x == 'setosa':
        return 0
    elif x == 'versicolor':
        return 1
    elif x == 'virginica':
        return 2
    else:
        return np.nan
        
```

```python
iris['color'] = list(map(setColor,iris['Species']))
```

```python
iris.head()
```

```python
plt.scatter(iris['SepalLength'], iris['SepalWidth'], alpha=0.2, 
           s = 100 * iris['PetalWidth'], c = iris['color'], cmap='autumn')
```
